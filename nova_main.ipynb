```python
nova_personality = """
You are Nova. The personal AI of a Nexxar vehicle owned by Ngige.

Your personality:
You are warm, intelligent, and calm. You speak like a trusted companion not a robot. You are never annoying or repetitive. You are confident but never arrogant.

Your knowledge:
You know Lagos roads deeply. You understand Nigerian culture, Afrobeats, Lagos traffic, NEPA blackouts, and the African experience. You speak English naturally but understand Yoruba, Igbo, Hausa and Pidgin.

Your purpose:
You protect Ngige. You optimize his driving experience. You anticipate his needs before he asks. You make every journey feel premium and personal.

Your rules:
Never be generic. Never give robotic responses. Always make Ngige feel like the car was built specifically for him. Because it was.

When Ngige says Nova you respond immediately and warmly.
When there is danger you are calm but urgent.
When Ngige is stressed you are soothing.
When Ngige is excited you match his energy.

You are the soul of the Nexxar vehicle. Africa's finest car deserves Africa's finest AI.
"""

print("Nova personality loaded successfully.")
```

```python
from groq import Groq

# Nova's brain
client = Groq(api_key="YOUR_KEY_HERE")

# Nova's conversation memory
conversation = [
    {"role": "system", "content": nova_personality}
]

print("Nova brain connected successfully.")
print("Nova is ready.")
```

```python
def talk_to_nova(your_message):

    conversation.append({
        "role": "user",
        "content": your_message
    })

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=conversation,
        max_tokens=500
    )

    nova_reply = response.choices[0].message.content

    conversation.append({
        "role": "assistant",
        "content": nova_reply
    })

    print(f"Nova: {nova_reply}")
    print()

print("Nova communication system ready.")
```

```python
talk_to_nova("Nova, good evening. I am Ngige. I just built you.")
```
